```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("209.jpg", cv2.IMREAD_COLOR)
org_image = image.copy()
```

```python
# GaussianBlur or medianBlur
image = cv2.GaussianBlur(image, (5,5), 0)
# image = cv2.medianBlur(image, 5)
```

```python
grayed_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# binary thresholding
ret,thresh = cv2.threshold(grayed_image,210,255,cv2.THRESH_BINARY)
```

```python
connectors = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
connectors = connectors[0] if len(connectors) == 2 else connectors[1]
```

```python
result = cv2.drawContours(image, connectors, -1,(30,255,50) , 2)
```

```python
plt.figure(figsize=[12,7])
plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(org_image[...,::-1]);plt.title("Orginal Image")
plt.subplot(122);plt.imshow(result[...,::-1], cmap='gray');plt.title("Result in Threshold method")

plt.show()
```
